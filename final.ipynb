#the knn model

the knn neighborsclassifieris is working on finding the best count on neigbors to have the best resaults, so i made a minimum maximum proggram to tell me which count of neighbors have the best results

```python
max_k = 2
knn3 = KNeighborsClassifier(n_neighbors = max_k)  
knn3.fit(X_train,Y_train)
max = knn3.score(X_test, Y_test)
print("k = {} score test: {}".format(max_k,max))  
print("\n"+"*"*15+"\n")

for k in range(3,8):
    knn3 = KNeighborsClassifier(n_neighbors = k)  
    knn3.fit(X_train,Y_train)
    score = knn3.score(X_test, Y_test)
    print("k = {} score test: {}".format(k,score))  
    print("\n"+"*"*15+"\n")
    time.sleep(1)
    if score > max:
      max = score 
      max_k = k

#just for organize
for i in range(5):
  if(i!=2):
    print("*"*15)
  else:
    print("\nconfusion matrix on the best score:\n")

print("max k = "+str(max_k))
knn3 = KNeighborsClassifier(n_neighbors = max_k)  
knn3.fit(X_train,Y_train)     
y_pred = knn3.predict(X_test)
score=accuracy_score(Y_test,y_pred)
print(score)
confusion_matrix(Y_test,y_pred)
matrix = classification_report(Y_test, y_pred)
print("the clasification report is:\n",matrix)
 
```

#tensorflow

```python
#elu sigmoid softmax relu
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(70, activation = "relu"),
    tf.keras.layers.Dense(80, activation = "sigmoid"),
    tf.keras.layers.Dense(100, activation = "softmax"),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(80, activation = "elu"),
    tf.keras.layers.Dense(60, activation = "relu"),
    tf.keras.layers.Dense(40, activation = "sigmoid"),
    tf.keras.layers.Dense(4, activation = "softmax")
    

])


loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
model.compile(optimizer='adam',
              loss=loss_fn,
              metrics=['accuracy'])


model.fit(X_train, Y_train, epochs = 10)
model.evaluate(X_test,  Y_test)

```
